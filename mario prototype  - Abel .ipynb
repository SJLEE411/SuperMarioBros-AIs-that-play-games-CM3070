```python
# MODEL LEARNINGReinforcement Learning
import os
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import BaseCallback

print("Stable-Baselines3 and PyTorch are installed correctly!")
```

```python
class TrainAndLoggingCallback(BaseCallback):

    def __init__(self, check_freq, save_path, verbose=1):
        super(TrainAndLoggingCallback, self).__init__(verbose)
        self.check_freq = check_freq
        self.save_path = save_path

    def _init_callback(self):
        if self.save_path is not None:
            os.makedirs(self.save_path, exist_ok=True)

    def _on_step(self):
        if self.n_calls % self.check_freq == 0:
            model_path = os.path.join(self.save_path, 'TESTING_WITH_RECORDING{}'.format(self.n_calls))
            self.model.save(model_path)

        return True
```

```python
CHECKPOINT_DIR ='./train/'
LOG_DIR ='./logs/'
```

```python
# every 10000 steps we get a new running for mario
callback = TrainAndLoggingCallback(check_freq=10000, save_path=CHECKPOINT_DIR)
```

```python
# creating AI model, AI mdoel starting
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=LOG_DIR, learning_rate=0.000001,
            n_steps=512)
```

```python
# learn model, train the AI model, 1000000 AI is goiong to see this many frames
model.learn(total_timesteps=1000000, callback=callback)
```

```python
import gym
import gym_super_mario_bros
from gym_super_mario_bros.actions import SIMPLE_MOVEMENT
from nes_py.wrappers import JoypadSpace
import torch
import numpy as np
from collections import deque
import cv2

# Convert the observation to grayscale and resize it
def preprocess_observation(obs):
    obs = cv2.cvtColor(obs, cv2.COLOR_RGB2GRAY)  # Convert to grayscale
    obs = cv2.resize(obs, (256, 240))  # Resize to match the expected size
    return obs

# Stack frames to capture temporal information
def stack_frames(stacked_frames, frame, is_new_episode):
    if is_new_episode:
        # Clear the stack and start with the current frame repeated 4 times
        stacked_frames = deque([frame] * 4, maxlen=4)
    else:
        # Append the new frame to the stack
        stacked_frames.append(frame)
    
    # Stack the frames along the channel dimension
    stacked_state = np.stack(stacked_frames, axis=0)
    return stacked_state, stacked_frames

# Initialize the environment and stack frames
env = gym_super_mario_bros.make('SuperMarioBros-1-1-v0')
env = JoypadSpace(env, SIMPLE_MOVEMENT)
stacked_frames = deque(maxlen=4)  # Initialize the frame stack

# Reset the environment and preprocess the first observation
state = env.reset()
state = preprocess_observation(state)
state, stacked_frames = stack_frames(stacked_frames, state, True)

while True:
    state = np.array(state)  # Ensure the state is a NumPy array
    state = state.reshape(-1, *state.shape[-2:])  # Reshape to (4, 240, 256)
    
    action, _state = model.predict(state)
    action = int(action)  # Convert action to an integer
    next_state, reward, done, info = env.step(action)
    
    # Preprocess and stack the next observation
    next_state = preprocess_observation(next_state)
    state, stacked_frames = stack_frames(stacked_frames, next_state, False)
    
    env.render()

    if done:
        # Reset the environment and stack frames for the next episode
        state = env.reset()
        state = preprocess_observation(state)
        state, stacked_frames = stack_frames(stacked_frames, state, True)
```
